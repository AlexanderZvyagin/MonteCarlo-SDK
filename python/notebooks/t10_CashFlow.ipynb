```python
from init import *
```

```python
model = sdk.Model()
model.TimeStart = 0
model.TimeSteps = 1000
model.NumPaths = 2
model.Add(sdk.IndependentGaussian())
```

```python
time_horizon = 2.5 # in years
```

```python
short_rate_start = 0.02
short_rate_process = model.Add(sdk.BrownianMotion(
    start     = short_rate_start,
    drift     = 0.0,
    diffusion = 0.0,
    title     = "short rate"
))
```

```python
discount_process = model.Add(sdk.ZeroCouponBond(
    underlying = short_rate_process.GetStateNumber(),
    start      = 1,
    title      = "discount factor"
))
```

```python
# CashFlows: every 0.5 year, till the time_horizon
CF_Ti = []
i = 1
dt = 0.5
while True:
    t = dt*i
    i += 1
    if t <= time_horizon:
        CF_Ti.append (t)
    else:
        break
print('CashFlow points:',CF_Ti)
```

```python
CashFlows = model.Add(sdk.CashFlows(
    underlying = discount_process.GetStateNumber(),
    t = CF_Ti,
    title = f'CashFlows of {discount_process.title}'
))
```

```python
dt = 0.25
i = 1
while True:
    t = dt*i
    i += 1
    if t <= time_horizon:
        model.evaluations.append(sdk.EvaluationPoint(0,t))
    else:
        break
```

```python
results = sdk.run (model, server)
results.df()
```
